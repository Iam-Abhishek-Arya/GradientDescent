```python
def gd(x_train,y_train,learning_rate,num_iterations):
    q = x_train.shape[1]
    m = np.zeros(q)
    for i in range(num_iterations):
        m = step_gradient(x_train,y_train,learning_rate,m)
        print(i,cost(x_train,y_train,m))
        #print(m)
    return m
```

```python
def step_gradient(x_train,y_train,learning_rate,m):
    q = x_train.shape[1]
    M = len(x_train)
    m_slopes = np.zeros(len(m))
    for i in range(M):
        for j in range(q):
            m_slopes[j] += (-2/M)*(y_train[i] - ((m*x_train[i,:]).sum())) * x_train[i,j]
            #m[j] = m[j] -  learning_rate * (-2/M)*(y_train[i] - ((m*x_train[i,:]).sum())) * x_train[i,j]      
    m =  m - learning_rate * m_slopes
    return m
```

```python
def cost(x,y,m):
    total_cost = 0
    M = len(x)
    for i in range(M):
        total_cost += (1/M) * (y[i] - (m * x[i,:]).sum())**2
    return total_cost
```

```python
def predict(x,m):
    y_pred = (x.dot(m[:-1]) + m[-1])
    return y_pred
```

```python
def score(y_pred,y_test):
    u = ((y_test - y_pred)**2).sum()
    v = ((y_test - y_test.mean())**2).sum()
    return 1 - u/v
```
